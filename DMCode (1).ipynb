```python
#WEEK - 2

#Date can be formatted to be of format YYYY-MM-DD , which is easier to write and test against later.
#dropping duplicate formatted dates (taking care of duplicate values)
df.drop_duplicates(subset="Formatted Date",inplace=True)
#temperatures and humidity, wind speeds can be identical, hence no dropping is required here.
```

```python
#missing values for pressure have been treated. We will now check for null/missing values in other attributes
df.dropna(how='any',inplace=True)
df.isnull().values.any()
```

```python
#As the value is false, there are no NaN values in the dataframe
#To treat errors and outliers, we will revisit all the graphs and take care of the features individually
#Outliers are in temperature, humidity, apparent temperature
# Date, Summary, Precip Type and Daily summary are categorical values with no outliers, they can only serve as labels
#Wind bearing and loud cover do not affect the outcome of class labels, and hence shall be dropped
q=df['Temperature (C)'].quantile(1)
q1=df['Temperature (C)'].quantile(0)
df=df[df['Temperature (C)']<q]
df=df[df['Temperature (C)']>q1]
sns.boxplot(x=df['Temperature (C)'])
```

```python
q=df['Humidity'].quantile(1)
q1=df['Humidity'].quantile(0)
df=df[df['Humidity']<q]
df=df[df['Humidity']>q1]
sns.boxplot(x=df['Humidity'])
```

```python
q=df['Pressure (millibars)'].quantile(1)
q1=df['Pressure (millibars)'].quantile(0)
df=df[df['Pressure (millibars)']<q]
df=df[df['Pressure (millibars)']>q1]
sns.boxplot(x=df['Pressure (millibars)'])
```

```python
#Pressure (millibars) has been taken care of. Pre-emptive classification using SVMs for class labeling :
from sklearn.svm import SVC
svm=SVC()
main_val=['Temperature (C)','Humidity','Wind Speed (km/h)','Visibility (km)','Pressure (millibars)']
X=df[main_val]
Y=df['Summary']
#splitting dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
svm.fit(X_train,Y_train) #fit SVM model     
print('Accuracy of SVM classifier for summary on training set: {:.2f}'.format(svm.score(X_train, Y_train)))
```

```python
print('Accuracy of SVM classifier for summary on test set: {:.2f}'.format(svm.score(X_test, Y_test)))
```

```python
Y=df['Precip Type']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
svm.fit(X_train,Y_train) #fit SVM model     
print('Accuracy of SVM classifier for Precipitation Type on train set: {:.2f}'.format(svm.score(X_train, Y_train)))
```

```python
print('Accuracy of SVM classifier for Precipitation Type on test set: {:.2f}'.format(svm.score(X_test, Y_test)))
```

```python
#CV for Summary
from sklearn import cross_validation
n=5
Y=df['Summary']
for i in range(n):
    # for each iteration, randomly hold out 20% of the data as CV set
    X_train,X_test,Y_train,Y_test=cross_validation.train_test_split(X,Y,test_size=0.3,random_state=1)
    # train model and make predictions
    svm.fit(X_train,Y_train)
    print('Fold: ',(i+1))
    print('Accuracy of SVM classifier for summary on train set: {:.2f}'.format(svm.score(X_train, Y_train)))
    print('Accuracy of SVM classifier for summary on test set: {:.2f}'.format(svm.score(X_test, Y_test)))
```

```python
#CV for Precipitation Type
from sklearn import cross_validation
n=5
Y=df['Precip Type']
for i in range(n):
    # for each iteration, randomly hold out 30% of the data as CV set
    X_train,X_test,Y_train,Y_test=cross_validation.train_test_split(X,Y,test_size=0.3,random_state=1)
    # train model and make predictions
    svm.fit(X_train,Y_train)
    print('Fold: ',(i+1))
    print('Accuracy of SVM classifier for summary on train set: {:.2f}'.format(svm.score(X_train, Y_train)))
    print('Accuracy of SVM classifier for summary on test set: {:.2f}'.format(svm.score(X_test, Y_test)))
```

```python
#Nested CV for Precipitaion type
from sklearn import cross_validation
outer_scores = []
inner_mean_scores=[]

# outer cross-validation
for j in range(3):
    X_train_outer,X_test_outer,Y_train_outer,Y_test_outer=cross_validation.train_test_split(X,Y,test_size=0.3,random_state=1)

    inner_scores=[]
    # inner cross-validation
    for i in range(3):
        # for each iteration, randomly hold out 30% of the data as CV set
        X_train_inner,X_test_inner,Y_train_inner,Y_test_inner=cross_validation.train_test_split(X_train_outer,Y_train_outer,test_size=0.3,random_state=1)
        # fit training data of inner CV
        svm.fit(X_train_inner,Y_train_inner)
        inner_scores.append(svm.score(X_test_inner,Y_test_inner))

    # calculate mean score for inner folds
    inner_mean_scores.append(np.mean(inner_scores))

    # fit the selected model to the training set of outer CV
    # for prediction error estimation
    svm.fit(X_train_outer,Y_train_outer)
    outer_scores.append(svm.score(X_test_outer,Y_test_outer))

# show the prediction error estimate produced by nested CV
print('Outer CV mean score: %.2f' % (np.mean(outer_scores)))
```

```python
for i in range(3):
    print('Inner CV mean scores: %.2f'%(inner_mean_scores[i]))
```
